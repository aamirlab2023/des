```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 12, 1000)
x = 1000/(1 + 999*np.exp(-0.9908*t))
tdot = np.linspace(0, 11, 12)
xdot = 1000/(1 + 999*np.exp(-0.9908*tdot))
print(tdot)
```

```python
plt.figure(figsize=(8, 8))
plt.plot(t, x)
plt.scatter(tdot, xdot, s=20, c='k')
plt.show()
```
